```python
class Solution:
    def smallestStringWithSwaps(self, s: str, pairs) -> str:
        from collections import deque
        pairs = set([tuple(sorted(p)) for p in pairs if p[0] != p[1]])
        q = deque([s,])
        seen = set([s])
        ans = s
        while q:
            print(q)
            node = q.popleft()
            if node < ans:
                ans = node
            for p in pairs:
                l,r = p
                news = node[:l]+node[r]+node[l+1:r]+node[l]+node[r+1:]
                print(news)
                if news not in seen:
                    seen.add(news)
                    q.append(news)
        return ans
```
