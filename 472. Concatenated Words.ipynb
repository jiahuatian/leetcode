```python
class Solution:
    def findAllConcatenatedWordsInADict(self, words):
        words.sort(key=len)
        words = [w for w in words if w!='']
        from collections import defaultdict
        bdict = defaultdict(set)
        def compose(word):
            if word == '':
                return True
            for bw in bdict[word[0]]:
                if bw in word:
                    w = word
                    word = word.replace(bw,'')
                    if compose(word):
                        return True
                    word = w
            return False
        ans = []
        for w in words:
            if compose(w):
                ans.append(w)
            else:
                bdict[w[0]].add(w)
        return ans
```
