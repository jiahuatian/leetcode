```python
class Solution:
    def suggestedProducts(self, products, searchWord: str):
        candidates = []
        for word in products:
            if word[0] == searchWord[0]:
                candidates.append(word)
        candidates.sort()
        ans = []
        if candidates:
            if len(candidates) >= 3:
                ans.append(candidates[:3])
            elif len(candidates) == 2:
                ans.append(candidates[:2])
            elif len(candidates) == 1:
                ans.append(candidates)
        else:
            ans = [[]for _ in range(len(searchWord))]
            return ans
        for i in range(1,len(searchWord)):
            n = len(candidates)
            for j in range(n):
                thiscan = candidates.pop(0)
                if i < len(thiscan) and thiscan[i] == searchWord[i]:
                    candidates.append(thiscan)
            candidates.sort()
            if candidates:
                if len(candidates) >= 3:
                    ans.append(candidates[:3])
                elif len(candidates) == 2:
                    ans.append(candidates[:2])
                elif len(candidates) == 1:
                    ans.append(candidates)
            else:
                ans.append([])
        return ans
```
