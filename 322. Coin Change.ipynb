```python
class Solution:
    def coinChange(self, coins, amount: int) -> int:
        coins.sort()
        if amount == 0:
            return 0
        dp = [0 for i in range(amount+1)]
        for i in range(1,amount+1):
            thislist = []
            for j in range(i):
                for k in range(len(coins)-1,-1,-1):
                    if (i-j)%coins[k] == 0 and dp[j] != -1:
                        thislist.append(dp[j]+(i-j)//coins[k])
            if thislist:
                dp[i] = min(thislist)
            else:
                dp[i] = -1
        return dp[amount]
```

```python
class Solution:
    def coinChange(self, coins, amount: int) -> int:
        def dp(i):
            if i == 0:
                return 0
            ans = float('inf')
            for coin in coins:
                if coin <= i:
                    ans = min(ans,dp(i-coin)+1)
            return ans
        return dp(amount)
```

```python
coins = [1,2,5]
amount = 100
obj = Solution()
obj.coinChange(coins,amount)
```
