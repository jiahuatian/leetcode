```python
class Solution:
    def longestIncreasingPath(self, matrix) -> int:
        from collections import deque
        directions = [(1,0),(-1,0),(0,1),(0,-1)]
        m,n = len(matrix),len(matrix[0])
        dp = [[[(i,j)] for j in range(n)]for i in range(m)]
        longest = 1
        seen = set()
        for i in range(m):
            for j in range(n):
                #if (i,j) not in seen:
                    q = deque([[(i,j),],])
                    seen.add((i,j))
                    while q:
                        path = q.popleft()
                        x,y = path[-1]
                        for xi,yi in directions:
                            if 0<=x+xi<m and 0<=y+yi<n and matrix[x+xi][y+yi] > matrix[x][y] and (x+xi,y+yi) not in seen:
                                seen.add((x+xi,y+yi))
                                q.append(path+[(x+xi,y+yi)])
                                known = path+[(x+xi,y+yi)]
                                for l,p in enumerate(known):
                                    px,py = p
                                    if len(known)-l > len(dp[px][py]):
                                        dp[px][py] = known[l:]
                            elif 0<=x+xi<m and 0<=y+yi<n and matrix[x+xi][y+yi] > matrix[x][y] and (x+xi,y+yi) in seen:
                                known = path+dp[x+xi][y+yi]
                                for l,p in enumerate(known):
                                    px,py = p
                                    if len(known)-l > len(dp[px][py]):
                                        dp[px][py] = known[l:]
                    longest = max(longest,len(dp[i][j]))
        print([[len(dp[i][j]) for j in range(len(dp[0]))]for i in range(len(dp))])
        return longest
```
